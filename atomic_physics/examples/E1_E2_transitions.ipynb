```python
import numpy as np
import atomic_physics
from scipy.constants import pi, c,epsilon_0 as eps0
from atomic_physics.ions.ba138 import Ba138
from atomic_physics.ions.ba137 import Ba137
```

```python
ion138 = Ba138(magnetic_field = 4.6e-4)
ion137 = Ba137(magnetic_field = 4.6e-4)
```

# Electric Dipole Rabi Frequncies

### $^{138}$Ba$^+$: 493 nm transition

```python
upper = ion138.get_state_for_MI_MJ(Ba138.P12, M_I=0, M_J=1/2)
lower = ion138.get_state_for_MI_MJ(Ba138.S12, M_I=0, M_J=1/2)

P = 1e-3
w0 = 30e-6
I0 = 2*P/w0**2
E0 = np.sqrt(2*I0/(c*eps0))

print(ion138.get_rabi_E1(lower, upper, E0)/(1e6)/(2*pi))
print(ion138.get_rabi_laser(lower, upper, E0)/(1e6)/(2*pi))
```

# Electric Quadrupole Rabi Frequencies

### Ba138: 1762 nm transition

```python
upper = ion138.get_state_for_MI_MJ(Ba138.D52, M_I=0, M_J=3/2)
lower = ion138.get_state_for_MI_MJ(Ba138.S12, M_I=0, M_J=-1/2)

P = 1e-3
w0 = 5e-6
I0 = 2*P/w0**2
E0 = np.sqrt(2*I0/(c*eps0))

omega = ion138.get_transition_frequency_for_states((upper, lower), relative=False)
k0 = omega/c

""" E0*k0/2 is the (l=2, m=+2) spherical tensor component of the quadrupole drive
if the laser's E-field, k-vector and bias B-field are all mutually orthogonal """
print(ion138.get_rabi_E2(lower, upper, E0*k0/2)/(1e6)/(2*pi))
print(ion138.get_rabi_laser(lower, upper, E0*k0/2)/(1e6)/(2*pi))
```
